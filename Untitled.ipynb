```python

# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# importing iris dataset
try:
    iris = pd.read_csv('iris.csv')
    print("data loaded successfully.")
except fileNotFoundError:
    Print("Error! File Not found.")
    exit()
```

```python
# Task 1: Explore the dataset
print (iris)
print("ðŸ“Œ First 5 rows of the dataset:")
print(iris.head())

print("\nðŸ“Œ Dataset info:")
print(iris.info())

print("\nðŸ“Œ Checking for missing values:")
print(iris.isnull().sum())

# Since no missing values, no cleaning needed
```

```python
#Basic Data Analysis
print("\nðŸ“Š Basic Statistics:")
print(iris.describe())

print("\nðŸ“Š Mean values grouped by species:")
print(iris.groupby('species').mean())
```

```python
#Data Visualization
# Scatter Plot: Sepal length vs petal length
plt.figure(figsize=(7,5))
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=iris)
plt.title('Scatter Plot: Sepal_Length vs Petal_Length')
plt.xlabel('Sepal_Length')
plt.ylabel('Petal_Length')
plt.tight_layout()
plt.savefig("scatter_sepal_petal.png")
plt.show()
```

```python
# 3. Histogram: Distribution of petal width
plt.figure(figsize=(7,5))
plt.hist(iris['petal_width'], bins=20, edgecolor='black')
plt.title('Histogram: Petal Width Distribution')
plt.xlabel('Petal_Width')
plt.ylabel('Frequency')
plt.tight_layout()
plt.savefig("histogram_petal_width.png")
plt.show()
```

```python
# Histogram: Distribution of petal_length
plt.figure(figsize=(7,5))
plt.hist(iris['petal_length'], bins=20, edgecolor='black')
plt.title('Histogram: Petal_length Distribution')
plt.xlabel('Petal_length')
plt.ylabel('Frequency')
plt.tight_layout()
plt.savefig("histogram_petal_length.png")
plt.show()
```

```python
#  Histogram: Distribution of sepal width
plt.figure(figsize=(7,5))
plt.hist(iris['sepal_width'], bins=20, edgecolor='black')
plt.title('Histogram: sepal_Width Distribution')
plt.xlabel('sepal_width')
plt.ylabel('Frequency')
plt.tight_layout()
plt.savefig("histogram_sepal_width.png")
plt.show()
```

```python
# Bar Chart: Average petal length per species
plt.figure(figsize=(7,5))
sns.barplot(x='species', y='petal_length', data=iris, errorbar=None)
plt.title('Bar Chart: Avg Petal Length per species')
plt.xlabel('species')
plt.ylabel('petal_Length (cm)')
plt.tight_layout()
plt.savefig("bar_chart_petal_length.png")
plt.show()
```

```python
# 1. Line Chart (sample index vs. sepal length)
plt.figure(figsize=(7,5))
plt.plot(iris.index, iris['sepal_length'], label='sepal_length')
plt.title('Line Chart: sepal_length Over Sample Index')
plt.xlabel('Sample Index')
plt.ylabel('sepal_length')
plt.legend()
plt.tight_layout()
plt.savefig("line_chart_sepal_length.png")
plt.show()
```

```python
# Line Chart (sample index vs. petal length)
plt.figure(figsize=(8,5))
plt.plot(iris.index, iris['petal_length'], label='petal_length')
plt.title('Line Chart: petal_length over Sample Index')
plt.xlabel('Sample Index')
plt.ylabel('petal_length')
plt.legend()
plt.tight_layout()
plt.savefig("line_chart_petal_length.png")
plt.show()
```
